```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# from sklearn.metrics import accuracy_score,precision_score,recall_score,fl_score,roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score,accuracy_score,precision_score,f1_score
```

```python
import pickle
```

```python
df=pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
# first we will load the dataset and clean it
```

```python
df=df.dropna() # this will handle missing values
```

```python
df=df.drop("customerID",axis=1) # this is useless colum we don need it
```

```python
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"],errors='coerce')
df=df.dropna()
```

```python
# now we have to encode categorical data
```

```python
le=LabelEncoder()
for col in df.select_dtypes(include='object'):
  df[col]=le.fit_transform(df[col])
```

```python
le # here encoding is done
```

```python
# feature engineering part
```

```python
# creating a new feature
df["AvgCharges"]=df["TotalCharges"]/df["tenure"]
```

```python
# mow accordingly we will split data and train models
x=df.drop("Churn",axis=1)
y=df["Churn"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# now we have to train logistic regression model
```

```python
from sklearn.model_selection import cross_val_score
lr=LogisticRegression()
lr.fit(X_train,y_train)
lr_cv=cross_val_score(lr,X_train,y_train,cv=5)
print("Logistic Regression cv Score",lr_cv.mean())
```

```python
# now we will try to tune the hyperparamters
p_grid={
    "n_estimators":[100,200],
    "max_depth":[5,10]

    }
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
grid=GridSearchCV(RandomForestClassifier(),p_grid,cv=5)
```

```python
grid.fit(X_train,y_train)
best_model=grid.best_estimator_
```

```python
print("Best Parameters :",grid.best_params_)
```

```python
# doing prediction
```

```python
pred = best_model.predict(X_test)
```

```python
print(pred)
```

```python
# evaluation
```

```python
print("Accuracy :",accuracy_score(y_test,pred))
print("precision :",precision_score(y_test,pred))
print("recall L",recall_score(y_test,pred))
print("f1 score :",f1_score(y_test,pred))
print("ROC AUC",roc_auc_score(y_test,pred))
```

```python
print("Confusion MAtrix",confusion_matrix(y_test,pred))
```

```python
print("classification report :",classification_report(y_test,pred))
```

```python
# save model
pickle.dump(best_model,open("churn_model.pkl","wb"))
```
