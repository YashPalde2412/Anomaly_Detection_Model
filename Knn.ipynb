```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=0.60,
    random_state=0
)
```

```python
# initializing centroids
```

```python
k=4
np.random.seed(42)

centroids=X[np.random.choice(X.shape[0],k,replace=False)]
```

```python
# making funtion to calculate distance
```

```python
def euclidean_distance(a,b):
  return np.sqrt(np.sum((a-b)**2))
```

```python
# trying to implement k means algorithm
```

```python
max_iters=100
for iteration in range(max_iters):
  clusters=[]
```

```python
# assigining clusters
for point in X:
  distances=[]
  # for centroid in centroids:
  #   distances.append(euclidean_distance(point,centroid))

for point in X:
  distances=np.linalg.norm(point-centroids,axis=1)
```

```python
cluster=np.argmin(distances)
```

```python
clusters.append(cluster)
```

```python
clusters=np.array(clusters)
```

```python
# storing old clusters
```

```python
# recalculating centroids
for iteration in range(100):
  old_centroids=centroids.copy()
  for i in range(k):
    points=X[cluster==i]
    if len(points)>1:
      centroids[i]=np.mean(points,axis=0)
    else:
      centroids[i]=old_centroids[i]
    if np.allclose(old_centroids,centroids):
      print("COverage at iteration ",iteration)
      break
```

```python
# visualization part
import matplotlib.pyplot as plt
import numpy as np
clusetrs=np.array(clusters)
plt.scatter(X[:,0],X[:,1],c=clusters,cmap='viridis')

plt.scatter(centroids[:,0],centroids[:,1],color="red",s=100)

plt.show()
```
